```python
def EM_Data(URL):
    http = urllib3.PoolManager()
    r = http.request('GET', URL)
    stus = r.status
    if stus == 200:
        dstr = r.data.decode()
    else:
        dstr = "No results"
    
    return dstr
```

```python
Sub Update_stk_price()
With Application
    .ScreenUpdating = False
    .Calculation = False
End With
Dim sht, shti As Worksheet
Set shti = ThisWorkbook.Sheets("index")
With ThisWorkbook.Sheets("raw_price")
    .Cells.ClearContents
    str_raw = EM_Data(URL_price)
    str_len = Len(str_raw)
    pos = InStr(1, str_raw, "[")
    poss = InStr(1, str_raw, "]")
    str_new = Mid(str_raw, pos + 1, str_len - pos - (str_len - poss) - 1)
    strr = Replace(str_new, """", "")
    r = 2
    c = 26
    Do While InStr(1, strr, ",") > 0
        If r Mod 100 = 0 Then
            DoEvents
            Application.StatusBar = r
        End If
        For c = 1 To 25
        pos = InStr(1, strr, ",")
            If pos = 0 Then
                .Cells(r, c) = strr
            Else
                .Cells(r, c) = Left(strr, pos - 1)
                strr = Mid(strr, pos + 1, 9999999)
            End If
            If c >= 4 Then
                If VarType(.Cells(r, c)) <> 5 Then .Cells(r, c) = 0
            End If
        Next
        p_pre = .Cells(r, 10)
        p_open = .Cells(r, 11)
        p_high = .Cells(r, 12)
        p_low = .Cells(r, 13)
        p_now = .Cells(r, 4)
        p_chg = .Cells(r, 5)
        p_chgpct = .Cells(r, 6)
        If p_pre > 0 And p_high > 0 And p_low > 0 And p_open > 0 Then
            .Cells(r, c + 0) = p_open / p_pre - 1
            .Cells(r, c + 1) = p_high / p_pre - 1
            .Cells(r, c + 2) = p_low / p_pre - 1
            .Cells(r, c + 3) = -Round(p_chgpct - .Cells(r, c + 1), 4)
        End If
        .Cells(r, 1) = Time()
        If p_now <= 0 Then strr = ""
        If Len(.Cells(r, 2)) = 6 And (Left(.Cells(r, 2), 1) = 2 Or Left(.Cells(r, 2), 1) = 3 Or Left(.Cells(r, 2), 1) = 9) Then
            .Rows(r).Delete
            r = r - 1
        End If
        If InStr(1, .Cells(r, 3), "S") > 0 Then
            .Rows(r).Delete
            r = r - 1
        End If
        r = r + 1
    Loop
    Application.StatusBar = ""
    .Columns(c).Resize(, 4).NumberFormatLocal = "0.0%"
    .Columns(1).NumberFormat = "[$-F400]h:mm:ss AM/PM"
    pos_c = WorksheetFunction.Match("price_now_col", shti.Rows(1), 0)
    cn = shti.Cells(99, pos_c).End(xlUp).Row - 1
    .Cells(1, 1).Resize(1, cn) = WorksheetFunction.Transpose(shti.Cells(2, pos_c).Resize(cn))
    .Range("N:U").Delete
    .UsedRange.AutoFilter
    '.Cells(1, 1).CurrentRegion.Copy sht.Cells(1, 1)
End With
With Application
    .ScreenUpdating = True
    .Calculation = xlCalculationAutomatic
End With
End Sub
```
